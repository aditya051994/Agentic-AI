```python
from pydantic import BaseModel
```

```python
class person(BaseModel):
    name: str
    age: int
    city:str
```

```python
person=person(name='Aditya',age=30,city='Mumbai')
person
```

```python
type(person)
```

```python
person1=person(name='Aditya',age=30,city=12)
print(person1)
```

```python
from typing import Optional
class Employee(BaseModel):
    name: str
    age: int
    city: Optional[str] = None
    indian:Optional[bool] = True 
```

```python
employee=Employee(name='Rahul',age=28,city='Delhi')
employee
```

```python
from typing import Optional
from typing import List
class Student(BaseModel):
    name: List[str]
    age: int
    marks:Optional[float]=None
```

```python
John=Student(name=['John','Willson'],age=20,marks=88.5)
John
```

```python
#Nested Model
class Address(BaseModel):
    city: str
    state: str
    country: str

class Employee(BaseModel):  
    name: str
    age: int
    address: Address
```

```python
employee1=Employee(name='Aditya',age=30,address=Address(city='Mumbai',state='Maharashtra',country='India'))
```

```python
employee1
```

```python
#Field
from pydantic import Field
class Employee(BaseModel):  
    name: str = Field(max_length=50, description="The name of the employee")
    price:int=Field(gt=0,lt=100,description="The price of the product")
```

```python
John=Employee(name='John',price=20)
John
```
