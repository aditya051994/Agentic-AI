## Context Caching

```python
import time
from google import genai
from google.genai.types import CreateCachedContentConfig,Content,Part

client=genai.Client()
client
```

```python
#Uploading file in google server
file_path=[
    'Data/apple 10-q q1 2024.pdf',
    'Data/apple 10-q q2 2024.pdf'
]
uploaded_files=[]
for path in file_path:
    file=client.files.upload(file=path)
    while file.state.name=='PROCESSING':
        time.sleep(2)
        file=client.files.get(name=file.name)

    uploaded_files.append(file) 
```

```python
uploaded_files
```

```python
#going to store file for 5 min in system
parts=[]
for f in uploaded_files:
    part=Part.from_uri(file_uri=f.uri,mime_type=f.mime_type)
    parts.append(part)

contents=[Content(role='user',parts=parts)]
```

```python
cache=client.caches.create(
    model='gemini-2.5-flash-lite',
    config=CreateCachedContentConfig(
        display_name='Apple Q1 Q2 2024 reports',
        system_instruction='You are fincial analysist.User these Appple Quaetery reports to answer',
        contents=contents,
        ttl='1800s'
    )
)
```
